```python
def parse_args(cfg):
    parser = argparse.ArgumentParser(description='Train keypoints network')
    # general
    parser.add_argument('--cfg',
                        help='experiment configure file name',
                        default=cfg,
                        type=str)

    args, rest = parser.parse_known_args()
    # update config
    update_config(args.cfg)

    # training
    parser.add_argument('--frequent',
                        help='frequency of logging',
                        default=config.PRINT_FREQ,
                        type=int)
    parser.add_argument('--gpus',
                        help='gpus',
                        type=str)
    parser.add_argument('--workers',
                        help='num of dataloader workers',
                        type=int)

    args = parser.parse_args("")

    return args
```

```python
def reset_config(config, args):
    if args.gpus:
        config.GPUS = args.gpus
    if args.workers:
        config.WORKERS = args.workers
```

```python
import glob
```

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)
```

```python
cfgs = glob.glob(f"experiments/mpii/*/*.yaml")
```

```python
for cfg in cfgs:
  args = parse_args(cfg)
  reset_config(config, args)
  # Setup model
  model = eval('models.'+config.MODEL.NAME+'.get_pose_net')(
      config, is_train=True
  )
  n_params = count_parameters(model)
  print(f"{cfg} Params: {n_params:,}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict
import pandas as pd
```

```python
TD0 = [0.00, 0.07, 6.78, 66.89, 78.15]
TD1 = [0.00, 18.87, 70.59, 79.34, 80.53]
```

```python
df_dict = defaultdict(list)
for td_key, td_data in zip(["TD0", "TD1"], [TD0, TD1]):
  for t, td_i in enumerate(td_data):
    df_dict["t"].append(t)
    df_dict["val"].append(td_i)
    df_dict["td"].append(td_key)
df = pd.DataFrame(df_dict)
```

```python
baseline_avg_acc = 83.25
```

```python
from matplotlib.lines import Line2D
```

```python
plt.figure(figsize=(8,3))
g = sns.lineplot(x="t", y="val", hue="td", data=df)
g.set_title("ResNet10 [MPII Validation Set]\nAvg Performance over all Joints")
g.set_xlabel("Timestep, t")
g.set_ylabel("Accuracy (%)")
handles, labels = g.get_legend_handles_labels()
baseline_line = Line2D([0], [0], label='Baseline', color='r', linestyle="--")
handles.append(baseline_line)
labels.append("Baseline")
plt.legend(handles, labels, loc="lower right")
plt.axhline(baseline_avg_acc, xmin=0, xmax=5, c="r", ls="--")
plt.show()
```
