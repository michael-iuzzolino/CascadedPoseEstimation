```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import cv2
from torchvision.utils import make_grid
```

```python
p = "/tmp/tmp_outputs.pt"
```

```python
data = torch.load(p)
```

```python
X = data["x"]
y = data["target"]
outputs = data["outputs"]
```

```python
X.shape[0]
```

```python
rand_i = 0
```

```python
def renorm(ele):
  return (ele - ele.min()) / (ele.max() - ele.min())
```

```python
def combine_heatmap_joints(X, resize=()):
  X_joined = X.sum(axis=0)
  if resize:
    X_joined = cv2.resize(X_joined, resize, interpolation=cv2.INTER_CUBIC)
  return X_joined
```

```python
Xi = X[rand_i].permute(1,2,0)
Xi = renorm(Xi).cpu().detach().numpy()
Xi = cv2.cvtColor(Xi, cv2.COLOR_BGR2RGB)
yi = y[rand_i].cpu().detach().numpy()
y_total = combine_heatmap_joints(yi, resize=Xi.shape[:2])
y_total = renorm(y_total)
rand_outputs = outputs[:,rand_i].cpu().detach().numpy()
Xi.shape, yi.shape, rand_outputs.shape
```

```python
plt.figure(figsize=(5,5))
plt.imshow(Xi)
plt.axis("off")
```

```python
plt.figure(figsize=(5,5))
plt.imshow(y_total, cmap="gray")
plt.axis("off")
```

```python
out_combos = []
for i, out_i in enumerate(rand_outputs):
  out_combo = combine_heatmap_joints(out_i, resize=Xi.shape[:2])
  out_combo = torch.tensor(out_combo).unsqueeze(dim=0)
  out_combo = renorm(out_combo)
  out_combos.append(out_combo)

grid_im = make_grid(out_combos, padding=15, pad_value=0.9).permute(1,2,0)

grid_im.shape
```

```python
plt.figure(figsize=(14,5))
plt.imshow(grid_im)
plt.axis("off")
```
